```python
import nltk
import pandas as pd
```

```python
text = "Madam Herta is a peerless gem. Madam Herta is an unrivaled genius. Madam Herta is an inimitable beauty."
print(text)
```

Tokenize (words)

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
tokens = nltk.word_tokenize(text)
print(tokens)
```

Tokenize (remove punctuation / special characters and numbers)

```python
import re

sample_clean = re.sub(r'[^a-zA-Z\s]', '', text)  # remove punctuation, numbers, special chars
tokens_clean = nltk.word_tokenize(sample_clean)
print(tokens_clean)
```

Tokenize (remove punctuation and all lowercase)

```python
sample_lower = re.sub(r'[^a-zA-Z\s]', '', text).lower()
tokens_lower = nltk.word_tokenize(sample_lower)
print(tokens_lower)
```

Tokenize (remove punctuation, all lowercase, remove stop words)

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

tokens_no_stop = [word for word in tokens_lower if word not in stop_words]
print(tokens_no_stop)
```

Tokenize (remove punctuation, all lowercase, remove stop words, with POS tagging)

```python
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

tokens_tagged = pos_tag(tokens_no_stop)
print(tokens_tagged)
```
