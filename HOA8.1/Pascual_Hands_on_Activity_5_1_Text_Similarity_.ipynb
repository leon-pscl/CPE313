Remove characters from the lyrics

```python
import re

newline = lambda x: re.sub('\n', ' ', x) # remove \n
alphanumeric = lambda x: re.sub('\w*\d\w*', ' ', x) # remove alphanumeric words

corpus = beatles.text.map(newline).map(alphanumeric)
corpus.head()
```

Create a document term matrix using Count Vectorizer with the stop words turned on to English

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words="english")
X = cv.fit_transform(corpus).toarray()

dt = pd.DataFrame(X, columns=cv.get_feature_names_out()).set_index(beatles.song)
dt.head()
```

```python
dt.shape
```

Count vecotrizer converts text into numbers by building a vocabulary of every unique word across all 178 songs. Then, for each song, it counts how many time each vocabulary word appears. The result is a 178 x 2026 matrix, which is something that can be used for classifiers and other tasks.

Define the cosine similarity calculation

```python
from numpy import dot
from numpy.linalg import norm

cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))
```

Cosine similarity measures the angle between two vectors. If two songs use very similar words, their vectors point in the same direction and the score is close to 1.0. If they share no words, the score is 0

Calculate all similarities and sort by the most similar

```python
similarity = [cosine(imagine, song) for song in X]
sorted(list(zip(similarity, beatles.song)), reverse=True)[1:6]
```

Create the TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

cv_tfidf = TfidfVectorizer(stop_words="english")
X_tfidf = cv_tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=cv_tfidf.get_feature_names_out()).set_index(beatles.song)
dt_tfidf.head()
```

TF-IDF works the same as count vecotrizer. However, instead of frequency count, it assigns a weight to each word. Words that appear a lot in one song but rarely in other songs get a high weight (these are distinctive)
Words that appear in many songs get a low weight (these are common and not helpful for comparison)

Calculate all similarities and sort by the most similar using the TF-IDF data

```python
similarity_tfidf = [cosine(imagine, song) for song in X_tfidf]
sorted(list(zip(similarity_tfidf, beatles.song)), reverse=True)[1:6]
```

This finds the top 5 songs most lyrically similar to "Imagine", using TF-IDF weights.

```python
import re
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from numpy import dot
from numpy.linalg import norm
```

The dataset to be used for this supplementary activity is the [20 newsgrops text dataset](https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html), which contains 18000 newsgroups posts on 20 topics.

```python
# Load dataset
newsgroups = fetch_20newsgroups()

```

```python
# Build a DataFrame
df = pd.DataFrame({'text': newsgroups.data,
                   'label': [newsgroups.target_names[t] for t in newsgroups.target]})

df.tail()
```

```python
df.shape
```

The loaded data was the training set

```python
# 2. Remove special characters
def clean(text):
    text = re.sub(r'\n', ' ', text)          # remove newlines
    text = re.sub(r'[^a-zA-Z\s]', ' ', text) # remove non-alpha chars
    text = re.sub(r'\s+', ' ', text).strip()  # collapse whitespace
    return text

corpus = df['text'].map(clean)
corpus.tail()
```

```python
# 3a. Count Vectorizer document-term matrix
cv = CountVectorizer(stop_words='english')
X_cv = cv.fit_transform(corpus).toarray()

dt_cv = pd.DataFrame(X_cv, columns=cv.get_feature_names_out())
dt_cv.index = df['label']
print("Count Vectorizer shape:", dt_cv.shape)
dt_cv.head()
```

```python
# 3b. TF-IDF Vectorizer document-term matrix
tfidf = TfidfVectorizer(stop_words='english')
X_tfidf = tfidf.fit_transform(corpus).toarray()

dt_tfidf = pd.DataFrame(X_tfidf, columns=tfidf.get_feature_names_out())
dt_tfidf.index = df['label']
print("TF-IDF shape:", dt_tfidf.shape)
dt_tfidf.head()
```

```python
# 4. Cosine similarity function
cosine = lambda v1, v2: dot(v1, v2) / (norm(v1) * norm(v2))

# Compare doc 1 against all others — Count Vectorizer
query = X_cv[1]
cv_sims = [(cosine(query, X_cv[i]), df['label'].iloc[i], corpus.iloc[i][:60])
           for i in range(len(X_cv)) if i != 1]  # skip doc 1 (comparing against itself)

print("=== Count Vectorizer: Top 5 similar to doc 1 ===")
for score, label, snippet in sorted(cv_sims, reverse=True)[:5]:
    print(f"  [{score:.4f}] ({label}) {snippet}...")
```

```python
# Compare doc 1 against all others — TF-IDF
query_tfidf = X_tfidf[1]
tfidf_sims = [(cosine(query_tfidf, X_tfidf[i]), df['label'].iloc[i], corpus.iloc[i][:60])
              for i in range(len(X_tfidf)) if i != 1]

print("=== TF-IDF: Top 5 similar to doc 1 ===")
for score, label, snippet in sorted(tfidf_sims, reverse=True)[:5]:
    print(f"  [{score:.4f}] ({label}) {snippet}...")
```

The results show that both vectorizers ranked the same subset of data that is most similar to the first post/entry in the dataset. However, the similarity scores differ a bit, with TF-IDF outputting lower similarity scores. This difference show the different approaches of the two vectorizers, with the count vectorizer solely basing the similarity based on the number of similar words, while the TF-IDF vectorizer bases on weights.
