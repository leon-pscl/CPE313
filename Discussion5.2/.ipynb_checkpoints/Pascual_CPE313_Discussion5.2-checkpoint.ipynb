## 1. Find a specific word on the twitter content and number of occurrences


Example: Happy


400



```python
#Configuration
search_word = 'players'#change this to any word
case_sensitive = False

#regex pattern
flags = 0 if case_sensitive else re.IGNORECASE
pattern_word = rf'\b{re.escape(search_word)}\b'

#Search in every tweet individually
matches_per_tweet = df['content'].astype(str).apply(
    lambda tweet: len(re.findall(pattern_word, tweet, flags=flags))
)

total_occurrences = matches_per_tweet.sum() #how many times the word appears (the word may appear multiple times in one tweet)
tweets_containing  = (matches_per_tweet > 0).sum() #the word appeared in how many tweets?

print(f"  Word searched : '{search_word}'")
print(f"  Case-sensitive: {case_sensitive}")
print(f"  Total occurrences : {total_occurrences:,}")
print(f"  Tweets containing : {tweets_containing:,}")
print("=" * 50)

#Show sample tweets that contain the word
print("\nSample tweets containing the word:")
sample = df[matches_per_tweet > 0][['sentiment', 'content']].head(5)
for i, row in sample.iterrows():
    #Highlight the word in output
    highlighted = re.sub(pattern_word,
                         lambda m: f"[[ {m.group()} ]]",
                         row['content'], flags=flags)
    print(f"  [{row['sentiment']}] {highlighted}") #highlight the specific word being searched
```

## 2. Find  a specific special character and number of occurrences

Example: *

               500



```python
#Configuration
search_char = '*' #this can be changed

#Build regex pattern
pattern_char = re.escape(search_char)

#Search
matches_per_tweet_char = df['content'].astype(str).apply(
    lambda tweet: len(re.findall(pattern_char, tweet))
)

total_char_occurrences = matches_per_tweet_char.sum()
tweets_with_char       = (matches_per_tweet_char > 0).sum()

print(f"  Special character searched: '{search_char}'")
print(f"  Total occurrences : {total_char_occurrences:,}")
print(f"  Tweets containing : {tweets_with_char:,}")
print("=" * 50)

#Show sample tweets
print("\nSample tweets containing the character:")
sample_char = df[matches_per_tweet_char > 0][['sentiment', 'content']].head(5)
for i, row in sample_char.iterrows():
    print(f"  [{row['sentiment']}] {row['content']}")
```

## 3. Find ALL numeric characters  and number of occurrences

  - 1 , 500

     2, 300



```python
#Recreate combined text from all tweets
all_text = ' '.join(df['content'].astype(str).tolist())

#Extract all digits from combined text
all_digits = re.findall(r'\d', all_text)

digit_counts = Counter(all_digits)

print("  Numeric Character Occurrences (0–9)")
print(f"  {'Digit':<10} {'Count':>10}")
print(f"  {'-'*10} {'-'*10}")
for digit in sorted(digit_counts.keys()):
    print(f"  {digit:<10} {digit_counts[digit]:>10,}")
print(f"  TOTAL digits found: {len(all_digits):,}")
```

What I did here is that I combined all tweets within the dataframe into one string of text. Then, I used regex to count the number of numeric characters.

## 4. Find ALL URLs and remove the URLs.

```python
import re

URL_PATTERN = re.compile(
    r'https?://[^\s<>"{}|\\^\[\]`]+'
    r'|www\.[^\s<>"{}|\\^\[\]`]+',
    flags=re.IGNORECASE
)

#BEFORE: DataFrame of tweets that contain URLs
df_with_urls = df[df['content'].astype(str).str.contains(URL_PATTERN)].copy()
df_with_urls = df_with_urls[['tweet_id', 'sentiment', 'content']].reset_index(drop=True)


print(f"  Total tweets containing URLs: {len(df_with_urls):,}")

print("\nBEFORE — First 7 tweets with URLs:\n")
for i, row in df_with_urls.head(7).iterrows():
    print(f"  [{i+1}] Sentiment : {row['sentiment']}")
    print(f"       Tweet     : {row['content']}")
    print()

#AFTER: Copy and remove URLs
df_urls_removed = df_with_urls.copy()
df_urls_removed['content'] = df_urls_removed['content'].apply(
    lambda tweet: re.sub(URL_PATTERN, '', tweet).strip()
)

print("=" * 60)
print("\nAFTER — First 7 tweets with URLs removed:\n")
for i, row in df_urls_removed.head(7).iterrows():
    print(f"  [{i+1}] Sentiment : {row['sentiment']}")
    print(f"       Tweet     : {row['content']}")
    print()
```

For this task, first thing I did was to isolate the tweets that have URLs into another dataframe. Then, I made a copy of that dataframe. In that copy, I used regex to remove the URLs within the tweets.

The result is one dataframe that has tweets w/ URLs, and another dataframe with the same tweets, but w/o URLs.
